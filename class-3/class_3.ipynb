### problem 4: find the factorial of a number

```python
n = 5
fact = 1
for i in range(1,n+1):
  fact *= i
print(fact)
```

### problem 5: print multiplication table of a number up to 10

```python
n = int(input('Put a number: '))
i =1
while i<=10:
  print(n,'x',i,'=',n*i)
  i += 1
```

### problem 6: reverse a list

```python
A =[36, 69, 24, 32, 45, 62, 45, 93]
i = 0
while i<= len(A)/2:
  A[i],A[len(A)-i-1] = A[len(A)-i-1],A[i]
  i += 1
print(A)
```

### problem 7: Sum of digit in a number

```python
sum_ = 0
num = 684
while num>0:
  sum_ += num%10
  num //= 10 # or num = int(n/10)
print(sum_)
```

### problem 8: find the maximum in a list

```python
A =[36, 69, 24, 32, 45, 62, 45, 93]
mx = A[0]
i = 0
while i<len(A):
  if A[i]>mx:
    mx = A[i]
  i += 1
print(mx)
```

# break/continue in loop

```python
while True:
  line = input('Put a line: ')
  if line == 'done':
    break
  print(line)
```
