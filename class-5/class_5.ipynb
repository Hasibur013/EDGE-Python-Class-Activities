```python
def Factorial():
  n = int(input('Put a number: '))
  fact = 1
  for i in range(1,n+1):
    fact *= i
  print(fact)

Factorial()
```

```python
# 1. write a python program to get 3 numbers from user to find gretest number

def greatestNumber():
  num1 = int(input('Enter num1:'))
  num2 = int(input('Enter num2:'))
  num3 = int(input('Enter num3:'))

  if num1>num2:
    if num1>num3:
      print('Greatest number: ', num1)
    else:
      print('Greatest number: ', num3)
  else:
    if num2>num3:
      print('Greatest number: ', num2)
    else:
      print('Greatest number: ', num3)


greatestNumber()
```

```python
# write a python program to get a number from user to display multiplication table for that number in rverse mode

def mult():
  num = int(input('put a number: '))

  while num>0:
    print('5'+'*'+str(num)+'='+str(num*5))
    num -= 1

mult()
```

```python
# 1. Write a python program to find area of circle and Area of squre
def area(r):
  print('Area of circle: ', 3.1416*r**2)
  print('Area of Squre: ', r**2)

length = int(input('Length: '))
area(length)
```

```python
def count_char(sen):
  count = 0
  for char in sen:
    count+=1

  print(count)
sen = input('Enter a sentences: ')
count_char(sen)
```

```python
def percentage(a,b,c):
  print('Percentage: ', float(a/(a+b+c))*300)
  print('Average: ', float((a+b+c)/3))

a = int(input('Math score: '))
b = int(input('Bangla Score: '))
c = int(input('English score: '))
percentage(a,b,c)
```

```python
def get_username():
  while True:
    username = input('Enter Username: ')
    if 8<= len(username) <= 20:
      if username.isalnum():
        print('Username Valid!')
        return username
      else:
        print("Username should only contain alphanumeric characters (no spaces or special characters).")
    else:
      print("Username length should be between 8 and 20 characters.")

get_username()
```
