# p1

```python
import pandas as pd


employees_large = pd.read_csv('employees_large.csv')


average_salary = employees_large.groupby('Department')['Salary'].mean()

highest_avg_salary_department = average_salary.idxmax()

print("Average Salary per Department:\n", average_salary)
print("\nDepartment with Highest Average Salary:", highest_avg_salary_department)
```

# p2

```python

df_sales = pd.read_csv('sales_large.csv')

total_sales = df_sales.groupby(['Region', 'Product'])['Sales'].sum()

highest_selling_product = df_sales.groupby('Product')['Sales'].sum().idxmax()

print("Total Sales per Product per Region:\n", total_sales)
print("\nProduct with Highest Total Sales:", highest_selling_product)
```

# p3

```python
import numpy as np

A = np.array([[2, 4, 6], [1, 3, 5], [7, 9, 11]])
B = np.array([[8, 5, 2], [7, 4, 1], [6, 3, 0]])

sum_matrix = A + B

elementwise_product = A * B

matrix_product = np.dot(A, B)

determinant_of_A = np.linalg.det(A)

print("Sum of A and B:\n", sum_matrix)
print("\nElement-wise Product:\n", elementwise_product)
print("\nMatrix Product:\n", matrix_product)
print("\nDeterminant of A:", determinant_of_A)
```

# p4

```python
random_data = np.random.uniform(0, 1, 10000)

mean_value = np.mean(random_data)
median_value = np.median(random_data)
std_dev = np.std(random_data)

min_value = np.min(random_data)
max_value = np.max(random_data)

count_in_range = np.sum(random_data > mean_value)

normalized_df = (random_data - mean_value) / (max_value - min_value)

print("Mean:", mean_value)
print("Median:", median_value)
print("Standard Deviation:", std_dev)
print("Min Value:", min_value)
print("Max Value:", max_value)
print("Count of values greater than the mean:", count_in_range)

print("Normalized Data", normalized_df)

```

```python

temperatures = [28, 32, 35, 31, 29, 30, 33, 36, 27, 25, 34, 30, 29, 37, 38, 26, 31, 35,
33, 32, 36, 34, 29, 28, 27, 35, 32, 30, 31, 33]

high_temperatures = [temp for temp in temperatures if temp > 30]


categorized_temps = [(temp, "Hot" if temp > 35 else "Warm" if temp > 30 else "Cool") for temp in temperatures]


avg_temp = sum(temperatures) / len(temperatures)
temp_differences = [temp - avg_temp for temp in temperatures]

print("Temperatures above 30°C:", high_temperatures)
print("\nCategorized Temperatures:", categorized_temps)
print("\nTemperature Differences from Average:", temp_differences)
```
