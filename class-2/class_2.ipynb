### 1. Area of circle

```python
PI = 3.14
radius = 10
area = PI * radius**2
print(area)
```

### 2. Swap variables without using third variable

```python
a = 10
b = 20
a,b = b,a
print(a,b)
```

### 3. Simple Interest Calculation

```python
p = 1000
r = 10
t = 2
simpleInterset = (p*r*t)/100
print(simpleInterset)
```

### 4. Temparature Conversation

```python
celcius = 37
fahrenheit = (celcius * 9/5) + 32
print(fahrenheit)
```

### 5. Rectangle perimeter and area

```python
length = 10
width = 20
perimeter = 2 * (length + width)
area = length * width
print('Perimeter: ', perimeter,'\nArea: ' , area)
```
